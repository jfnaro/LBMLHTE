```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
raw_abalone = pd.read_csv('abalone.data', header=None, names=['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings'])
raw_abalone
```

```python
ages = raw_abalone['Rings'] * 1.5
raw_abalone.drop('Rings', axis=1, inplace=True)
ages
```

```python
raw_abalone = raw_abalone.join(pd.get_dummies(raw_abalone['Sex']))
raw_abalone.drop('Sex', axis=1, inplace=True)
raw_abalone
```

```python
abalone_train, abalone_test, age_train, age_test = train_test_split(raw_abalone, ages, test_size=0.2)
```

```python
from sklearn import linear_model
```

```python
type(np.array(age_train))
```

```python
# percep = linear_model.Perceptron().fit(abalone_train, pd.DataFrame(age_train))
# percep.score(abalone_test, age_test)
```

```python
lin_reg = linear_model.LinearRegression().fit(abalone_train, age_train)
lin_reg.score(abalone_test, age_test)
```

```python
SGD_reg = linear_model.SGDRegressor().fit(abalone_train, age_train)
SGD_reg.score(abalone_test, age_test)
```

```python
ARD_reg = linear_model.ARDRegression().fit(abalone_train, age_train)
ARD_reg.score(abalone_test, age_test)
```

```python
bayes_reg = linear_model.BayesianRidge().fit(abalone_train, age_train)
bayes_reg.score(abalone_test, age_test)
```
